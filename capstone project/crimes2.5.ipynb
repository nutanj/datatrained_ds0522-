```python
sns.relplot(x='population',y='Total IPC crimes',data=df)
plt.show()
```

```python
sns.regplot(x='population',y='Total IPC crimes',data=df)
plt.show()
```

in this regression plot it can be observed that, there is a upward pattern in the population and total crimes relation.
