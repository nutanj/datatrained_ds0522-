```python
sns.relplot(x='area',y='Total IPC crimes',data=df)
plt.show()
```

```python
sns.regplot(x='area',y='Total IPC crimes',data=df)
plt.show()
```

```python
plt.scatter(df['area'],df['Total IPC crimes'])
plt.show()
```

from this scatter plot we can conclude that, more crime areas are concentrated in the range 150000km area. again from these graphs we can assume that, larger states have high number of crimes. rajastan is the largest state area wise and it has a huge criminal incidents reported. but at the same time, we can see that maharastra has the most number of crimes committed. the union territories like daman and diu and lakshyadweep has low numbers of crimes.
